## A - Meal Delivery

```
x, a, b = map(int, input().split())
ans = 'A' if abs(a - x) < abs(b - x) else 'B'
print(ans)
```

## B - Not Found

```
s_set = set(list(input()))
alphabet_set = set('abcdefghijklmnopqrstuvwxyz')

ans = 'None'

for alphabet in alphabet_set:
  if alphabet not in s_set:
    ans = alphabet
    break

print(ans)
```

## C - Make a Rectangle

```
from collections import defaultdict
n = int(input())
a_list = list(map(int, input().split()))

a_dict = defaultdict(int)

for a in a_list:
  a_dict[a] += 1

edge_dict = dict(filter(lambda item: item[1] >= 2, a_dict.items()))

if len(edge_dict.items()) < 2:
  ans = 0
else:
  edge_dict = sorted(edge_dict.items(), reverse=True)
  if edge_dict[0][1] >= 4:
    ans = edge_dict[0][0] ** 2
  else:
    ans = edge_dict[0][0] * edge_dict[1][0]

print(ans)
```
