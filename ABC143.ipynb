## A - Curtain

```
a, b = map(int, input().split())
ans = max(0, a - 2 * b)
print(ans)
```

## B - TAKOYAKI FESTIVAL 2019

```
n = int(input())
d_list = list(map(int, input().split()))

ans = 0

for i in range(n - 1):
  for j in range(i + 1, n):
    ans += d_list[i] * d_list[j]

print(ans)
```

## C - Slimes

```
n = int(input())
s = input()

slime = s[0]
ans = 1

for i in range(1, n):
  if s[i] != slime:
    ans += 1
    slime = s[i]

print(ans)
```

## D - Triangles

```
# TLE
from operator import mul
from functools import reduce


def combination_cnt(n, r):
  # Combinationを算出 その2
  # 5C2 = 10など

  r = min(n - r, r)
  if r == 0: return 1
  return reduce(mul, range(n - r + 1, n + 1)) // reduce(mul, range(1, r + 1))


n = int(input())
l_list = list(map(int, input().split()))
l_list.sort()

_ans = 0

for i in range(n - 2):
  for j in range(i + 1, n - 1):
    if l_list[i] + l_list[j] > l_list[-1]:
      continue
    else:
      for k in range(j + 1, n):
        if l_list[i] + l_list[j] <= l_list[k]:
          _ans += n - k
          break

ans = combination_cnt(n, 3) - _ans
print(ans)
```

```
import bisect
from operator import mul
from functools import reduce


def combination_cnt(n, r):
  # Combinationを算出 その2
  # 5C2 = 10など

  r = min(n - r, r)
  if r == 0: return 1
  return reduce(mul, range(n - r + 1, n + 1)) // reduce(mul, range(1, r + 1))


n = int(input())
l_list = list(map(int, input().split()))
l_list.sort()

_ans = 0

for i in range(n - 2):
  for j in range(i + 1, n - 1):
    if l_list[i] + l_list[j] > l_list[-1]:
      continue
    else:
      _ans += n - bisect.bisect_left(l_list, l_list[i] + l_list[j])
      
ans = combination_cnt(n, 3) - _ans
print(ans)
```
