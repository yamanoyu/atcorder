## A - Two Sequences 2

```
n = int(input())
a_list= list(map(int, input().split()))
b_list= list(map(int, input().split()))

a_max = a_list[0]
ans = a_list[0] * b_list[0]

print(ans)
for i in range(1, n):
  a_max = max(a_max, a_list[i])
  ans = max(ans, a_max * b_list[i])
  print(ans)
```

## B - Mex Boxes

```
from collections import defaultdict


n, k = map(int, input().split())
a_list= list(map(int, input().split()))

a_dict = defaultdict(int)

for a in a_list:
  a_dict[a] += 1

a_sort_dict = dict(sorted(a_dict.items(), key=lambda x: x[0]))

ans = 0
num_cnt = k
num = -1

for a_key, a_val in a_sort_dict.items():
  if a_key == num + 1:
    if a_val < num_cnt:
      ans += a_key * (num_cnt - a_val)
      num_cnt = a_val
    num += 1
  else:
    ans += (num + 1) * num_cnt
    num_cnt = 0
    break

if num_cnt >= 1:
  ans += (num + 1) * num_cnt

print(ans)
```
