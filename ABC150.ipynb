## A - 500 Yen Coins

```
k, x = map(int, input().split())
ans = 'Yes' if 500 * k >= x else 'No'
print(ans)
```

## B - Count ABC

```
n = int(input())
s = input()
ans = 0
for i in range(n - 2):
  if s[i:i+3] == 'ABC':
    ans += 1
print(ans)
```

## C - Count Order

```
import itertools


n = int(input())
p_list = list(map(int, input().split()))
q_list = list(map(int, input().split()))

p_str = ''.join(list(map(str, p_list)))
q_str = ''.join(list(map(str, q_list)))

permutations_list = itertools.permutations(range(1, n + 1), n)
permutations_str_list = [''.join(list(map(str, p))) for p in permutations_list]

ans = abs(permutations_str_list.index(p_str) - permutations_str_list.index(q_str))
print(ans)
```

## D - Semi Common Multiple

```
import math


n, m = map(int, input().split())
a_list= list(map(int, input().split()))

lcm = a_list[0] // 2

for i in range(1, n):
  lcm *= (a_list[i] // 2) // math.gcd(lcm, a_list[i] // 2)

f = 0

for a in a_list:
  if (lcm // (a // 2)) % 2 == 0:
    f = 1

ans = (m // lcm + 1) // 2 if f == 0 else 0
print(ans)
```
